# Feature Engineering: Finding Optimal Decay Parameters for Score-Based Features

## Introduction

In this notebook, we explore how to choose decay parameters for the categorical score features introduced in our modeling pipeline—specifically for directors, writers, and cast members. Each score is designed to capture the past performance of a person or company based on the vote averages of their previous movies, with more recent work weighted more heavily using exponential decay. The precise formulas used to compute these scores are given in the appropriate sections below, but they all depend on one or more decay parameters (e.g., for time or cast order) that we aim to tune here.

To determine appropriate decay parameters, we take a simple, visual approach:  
- For each feature (e.g., director score), we vary the decay parameter $\lambda$ over a range of values.  
- For each $\lambda$, we compute the correlation between the resulting scores and the vote average of the current movie.  
- We then plot this relationship and visually inspect the plot to identify the value of $\lambda$ that maximizes the correlation.

This method is not guaranteed to yield globally optimal parameters—especially since the best $\lambda$ for one score may not work well when combined with the others in the final model. However, for our purposes, this approach gives us a reasonable approximation of effective decay values that encode quality and recency in a meaningful way.

```python
import sqlite3
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from config.settings import DB_PATH
```

## Director Score

To quantify the impact of a director's past work on the success of a movie, we compute a **director score** based on the vote averages of their previous films, giving more weight to recent movies using exponential decay. The formula for the director score is as follows:

$$\text{Director Score} = \frac{\sum\limits_{m} \text{vote\_ave}_m \cdot e^{-\lambda \cdot \text{days}_m} }{\sum\limits_{m} e^{-\lambda \cdot \text{days}_m}}$$

Here:

- $\text{vote\_ave}_m$ is the vote average for the previous movie $m$.
- $\text{days}_m$ is the number of days between the release of previous movie $m$ and the release of the current movie being scored.
- $\lambda$ is the decay parameter that controls how much recent movies are weighted more heavily than older ones.

This formula provides a weighted average of the director’s past work, allowing us to incorporate both the quality and recency of their films into the score.


```python
director_query = """
WITH past_movie AS (
    SELECT 
        m.movie_id,
        pm.vote_average,
        pm.release_date AS prev_release_date,
        m.release_date AS current_release_date,
        JULIANDAY(m.release_date) - JULIANDAY(pm.release_date) AS days_difference
    FROM movie m
    JOIN movie_crew mc1 ON m.movie_id = mc1.movie_id  
    JOIN movie_crew mc2 ON mc1.person_id = mc2.person_id  
    JOIN movie pm ON mc2.movie_id = pm.movie_id  
    WHERE mc1.job IN ('Director', 'Co-Director') 
      AND mc2.job IN ('Director', 'Co-Director')
      AND pm.release_date < m.release_date
      AND pm.vote_count >= 30  
)
SELECT 
    m.movie_id, 
    m.vote_average,
    COALESCE(
        SUM(pm.vote_average * EXP(-? * days_difference)) / 
        SUM(EXP(-? * days_difference)), 
        (SELECT AVG(vote_average) FROM movie WHERE vote_count >= 30)  
    ) AS director_score
FROM movie m
LEFT JOIN past_movie pm ON m.movie_id = pm.movie_id
WHERE m.vote_count >= 30
GROUP BY m.movie_id;
"""
```

```python
def plot_correlations(lambda_values, query, score):
    correlations = []  # Store correlation results

    with sqlite3.connect(DB_PATH) as conn:
        # Loop through lambda values
        for lambda_val in lambda_values:
            df = pd.read_sql_query(query, conn, params=[lambda_val, lambda_val])
            
            # Compute correlation
            corr = df["vote_average"].corr(df[score])
            correlations.append(corr)
    
    # Plot the results
    plt.figure(figsize=(8, 5))
    plt.plot(lambda_values, correlations, marker='o', linestyle='-')
    plt.xscale("log")  # Log scale for better visualization
    plt.xlabel("Lambda (Time Decay Factor)")
    plt.ylabel(f"Correlation between {score} and Vote Average")
    plt.title(f"Optimal Lambda Selection for {score}")
    plt.grid(True)
    plt.show()
```

```python
lambda_values = np.logspace(-7, 1, 10)  # initial lambda values to test
plot_correlations(lambda_values, director_query, "director_score")
```

It looks like setting $\lambda=4\times 10^{-4}=0.0004$ will roughly optimize the correlation between the director score and the vote average. 

## Writers Score

The **writers score** follows the same structure as the director score, aiming to capture the recent performance of writers based on the vote averages of their previous films. As with directors, we apply exponential decay to give more weight to recent movies:

$$\text{Writer Score} = \frac{\sum\limits_{m} \text{vote\_ave}_m \cdot e^{-\lambda \cdot \text{days}_m} }{\sum\limits_{m} e^{-\lambda \cdot \text{days}_m}}$$

Where:

- $\text{vote\_ave}_m$ is the vote average of the previous movie $m$ written by the writers.
- $\text{days}_m$ is the number of days between the release of the previous movie $m$ and the current movie.
- $\lambda$ is the decay parameter controlling how quickly older movies are discounted.

This approach mirrors the director score and serves the same purpose: to summarize the writers' recent track record in a way that reflects both quality and recency.


```python
writers_query = """
WITH past_movie AS (
    SELECT 
        m.movie_id,
        pm.vote_average,
        pm.release_date AS prev_release_date,
        m.release_date AS current_release_date,
        JULIANDAY(m.release_date) - JULIANDAY(pm.release_date) AS days_difference
    FROM movie m
    JOIN movie_crew mc1 ON m.movie_id = mc1.movie_id  
    JOIN movie_crew mc2 ON mc1.person_id = mc2.person_id  
    JOIN movie pm ON mc2.movie_id = pm.movie_id  
    WHERE mc1.department = 'Writing' 
      AND mc2.department = 'Writing'
      AND pm.release_date < m.release_date
      AND pm.vote_count >= 30  
)
SELECT 
    m.movie_id, 
    m.vote_average,
    COALESCE(
        SUM(pm.vote_average * EXP(-? * days_difference)) / 
        SUM(EXP(-? * days_difference)), 
        (SELECT AVG(vote_average) FROM movie WHERE vote_count >= 30)  
    ) AS writers_score
FROM movie m
LEFT JOIN past_movie pm ON m.movie_id = pm.movie_id
WHERE m.vote_count >= 30
GROUP BY m.movie_id;
"""
```

```python
lambda_values = np.logspace(-7, 1, 10)  # initial lambda range
plot_correlations(lambda_values, writers_query, "writers_score")
```

```python
lambda_values = np.logspace(-4, -3, 10)  # Adjusting lambdas based on graph above
plot_correlations(lambda_values, writers_query, "writers_score")
```

```python
lambda_values = np.logspace(-3.5, -3.1, 10)  # Adjusting lambdas based on graph above
plot_correlations(lambda_values, writers_query, "writers_score")
```

### Choice of $\lambda$ for Writers

It looks like setting $\lambda=4\times 10^{-4}=0.0004$, the same as the value for the director score, will also roughly optimize the correlation between the writers score and the vote average. 

## Cast Score

The **cast score** is similar in spirit to the director and writer scores, aiming to reflect the recent performance of the movie's cast based on their prior work. However, because a movie typically has multiple cast members with varying prominence, we modify the formula to include both **recency** and **cast order** (i.e., billing position).

The formula for the cast score is:

$$
\text{Cast Score} = \frac{\sum\limits_{m} \text{vote\_ave}_m \cdot e^{-\lambda_t \cdot \text{days}_m - \lambda_o \cdot \text{cast\_order}_m}}{\sum\limits_{m} e^{-\lambda_t \cdot \text{days}_m - \lambda_o \cdot \text{cast\_order}_m}}
$$

Where:

- $\text{vote\_ave}_m$ is the vote average of a previous movie $m$ that included the cast member.
- $\text{days}_m$ is the number of days between the release of movie $m$ and the current movie.
- $\text{cast\_order}_m$ indicates the cast member’s billing position in movie $m$ (lower is more prominent).
- $\lambda_t$ and $\lambda_o$ are decay parameters for time and cast order, respectively.

This formulation gives higher weight to performances that are both more recent and from more prominent roles, producing a nuanced summary of cast quality relevant to the current movie.


```python
cast_query = """
WITH RelevantCast AS (
    -- Get top 10 cast members per movie (or fewer if there are fewer cast members)
    SELECT mc.movie_id, mc.person_id, mc.cast_order, m.release_date, m.vote_average
    FROM movie_cast mc
    JOIN movie m ON mc.movie_id = m.movie_id
    WHERE mc.cast_order <= 10 -- Consider only the top 10 billed actors
),
CastHistory AS (
    -- Join movies with their past appearances of cast members
    SELECT 
        m.movie_id,
        prev_m.vote_average AS past_vote_average,
        JULIANDAY(m.release_date) - JULIANDAY(prev_m.release_date) AS days_difference,
        rc.cast_order
    FROM movie m
    JOIN RelevantCast rc ON m.movie_id = rc.movie_id
    JOIN movie_cast prev_mc ON rc.person_id = prev_mc.person_id
    JOIN movie prev_m ON prev_mc.movie_id = prev_m.movie_id
    WHERE prev_m.release_date < m.release_date -- Only use past movies
    AND prev_m.vote_count >= 30 -- Only consider past movies with enough votes
)
SELECT 
    m.movie_id,
    m.vote_average,
    CASE 
        WHEN SUM(EXP(-? * days_difference) * EXP(-? * cast_order)) = 0 
        THEN NULL
        ELSE SUM(past_vote_average * EXP(-? * days_difference) * EXP(-? * cast_order)) / 
             SUM(EXP(-? * days_difference) * EXP(-? * cast_order))
    END AS cast_score
FROM CastHistory
JOIN movie m ON CastHistory.movie_id = m.movie_id
WHERE m.vote_count >= 30
GROUP BY m.movie_id;
"""
```

```python
def plot_correlation_heat_map(lambda_times, lambda_orders):
    # Initialize a DataFrame to store correlations
    correlation_matrix = np.zeros((len(lambda_times), len(lambda_orders)))
    
    # Loop over the grid of lambda values and compute correlations
    for i, lambda_time in enumerate(lambda_times):
        for j, lambda_order in enumerate(lambda_orders):
            # Execute query with parameters
            df = pd.read_sql_query(cast_query, conn, params=[lambda_time, lambda_order, lambda_time, lambda_order, lambda_time, lambda_order])
            
            # Compute correlation between cast_score and vote_average
            correlation = df['cast_score'].corr(df['vote_average'])
            
            # Store the result in the matrix
            correlation_matrix[i, j] = correlation
    
    # Create a heatmap of the correlation matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, xticklabels=lambda_orders, yticklabels=lambda_times, cmap='coolwarm', annot=True, fmt=".2f")
    plt.xlabel('lambda_order')
    plt.ylabel('lambda_time')
    plt.title('Correlation between cast_score and vote_average for different lambda_time, lambda_order pairs')
    plt.show()
```

```python
# Set up lambda values for grid search using logarithmic spacing
lambda_times = np.logspace(-6, -2, 5)  # initial lambda_time range
lambda_orders = np.logspace(-7, 0, 5)  # initial lambda_order range
plot_correlation_heat_map(lambda_times, lambda_orders)
```

## Production Company Score

Unlike the previous categorical scores, the **production company score** is computed more simply: we take the average vote score of all past movies produced by each company. We don't incorporate decay parameters like recency or prominence here, so there's no tuning step involved.

To evaluate how useful this score might be, we compute its correlation with the vote average of the current movie. This gives us a quick sense of whether a production company’s historical performance is predictive of a movie’s quality.

```python
production_co_query = """
WITH PastMovies AS (
    -- Get past movies with the same production company
    SELECT 
        m.movie_id,
        prev_mpc.company_id,
        prev_m.vote_average
    FROM movie m
    JOIN movie_production_company mpc ON m.movie_id = mpc.movie_id
    JOIN movie_production_company prev_mpc ON mpc.company_id = prev_mpc.company_id
    JOIN movie prev_m ON prev_mpc.movie_id = prev_m.movie_id
    WHERE prev_m.release_date < m.release_date
    AND prev_m.vote_count >= 30
),
CompanyAverages AS (
    -- Compute the average vote_average per production company
    SELECT 
        company_id,
        AVG(vote_average) AS company_avg_vote
    FROM PastMovies
    GROUP BY company_id
),
OverallAverage AS (
    -- Compute the overall average vote_average for all movies with enough votes
    SELECT AVG(vote_average) AS overall_avg_vote
    FROM movie
    WHERE vote_count >= 30
)
SELECT 
    m.movie_id,
    m.vote_average,
    COALESCE(company_avg_vote, overall_avg_vote) AS production_company_score
FROM movie m
LEFT JOIN movie_production_company mpc ON m.movie_id = mpc.movie_id
LEFT JOIN CompanyAverages ca ON mpc.company_id = ca.company_id
CROSS JOIN OverallAverage
WHERE m.vote_count >= 30
GROUP BY m.movie_id;
"""
```

```python
with sqlite3.connect(DB_PATH) as conn:
    df = pd.read_sql_query(production_co_query, conn)
print(df.head())
```

```python
print(f"Correlation between production company score and vote average: {df['production_company_score'].corr(df['vote_average']): .2f}")
```
